```python
import bibtexparser
import matplotlib.pyplot as plt

def analyze_bib_file(bib_text):
    """Analyzes a BibTeX string, extracts citations, and plots their distribution over years."""

    db = bibtexparser.loads(bib_text)

    years = {}
    for entry in db.entries:
        if 'year' in entry:
            year = int(entry['year'])
            years[year] = years.get(year, 0) + 1

    years_sorted = sorted(years.items())

    plt.figure(figsize=(10, 6))
    plt.bar([x[0] for x in years_sorted], [x[1] for x in years_sorted])
    plt.xlabel("Year")
    plt.ylabel("Number of Citations")
    plt.title("Citation Distribution Over Years")
    plt.xticks(rotation=45)
    plt.show()

# Inline .bib text for testing
bib_text = """
@article{smith2023,
  author = {Smith, John and Doe, Jane},
  title = {A Groundbreaking Study},
  journal = {Journal of Amazing Research},
  year = {2023},
  volume = {15},
  number = {2},
  pages = {100-125}
}
@book{jones2020,
  author = {Jones, Emily},
  title = {The Definitive Guide},
  publisher = {Academic Press},
  year = {2020}
}
@inproceedings{brown2018,
  author = {Brown, Michael},
  title = {Innovative Approach},
  booktitle = {Proceedings of the International Conference},
  year = {2018}
}
"""

# If you use your own bib file - make sure it is in the same directory as this notebook (ipynb file)
with open('references.bib', encoding='utf-8') as bibtex_file:
    bib_text = bibtex_file.read()

analyze_bib_file(bib_text)
```
